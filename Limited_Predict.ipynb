```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd  
import seaborn as sns 
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_boston
```

```python
#reading csv file
boston_dataset = pd.read_csv("boston.csv")
```

```python
#loading data in dataframe
X = pd.DataFrame(boston_dataset[['crim','zn','indus','chas','nox','rm','age','dis','rad','tax','ptratio','black','lstat']], columns =['crim','zn','indus','chas','nox','rm','age','dis','rad','tax','ptratio','black','lstat'])
Y = boston_dataset['medv'] 
print(X)
print(Y)
```

```python
#normalizing values
bost=(X-X.mean())/(X.max()-X.min())
bost.describe()
```

```python
#checking number of null values in each column
bost.isnull().sum()
```

```python
#plotting distribution of price
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.distplot(Y, bins=30)
plt.show()
```

```python
#using correlation matrix ,finding predictor variable
correlation_matrix = boston_dataset.corr().round(2)
sns.heatmap(data=correlation_matrix, annot=True)
```

```python
#plotting between predictor variable and price variable
plt.figure(figsize=(20, 5))

#list of predictor variables
features = [ 'rm','lstat','ptratio','black'] 
target = Y
```

```python
plt.figure(figsize=(20, 5))
for i, col in enumerate(features):
    plt.subplot(1, len(features) , i+1)
    x = bost[col]
    y = target
    plt.scatter(x, y, marker='o')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('MEDV')
```

```python
#removing unnecessary elements that do not affect price , and then storing in 'X'
X = pd.DataFrame(np.c_[bost['rm'],bost['ptratio'],bost['black'],bost['lstat']], columns = ['RM','PTRATIO','B','LSTAT'])
```

```python
#splitting data into train and test set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=5) 

# loading index values of randomly splitted dataframe
trindex=list(X_train.index)

# loading index values of randomly splitted dataframe
teindex=list(X_test.index) 
X_train=np.array(X_train)
Y_train=np.array(Y_train)
X_test=np.array(X_test)
Y_test=np.array(Y_test)
X_test1=X_test
```

```python
#shape of array
print("shape of X_train is: ",X_train.shape)
print("shape of X_test: ",X_test.shape)
print("shape of Y_train: ",Y_train.shape)
print("shape of Y_test: ",Y_test.shape)
```

```python
# creating object of LinearRegression
lin = LinearRegression() 

# fitting line in given data
lin.fit(X_train,Y_train) 
print("value of coefficients of line are: ",lin.coef_)
print("vaue of intercept of line is: ",lin.intercept_)
```

```python
# calculating average percentage error in predicted price for train set
per=[]
import math
for i in range(Y_train.shape[0]):
    Val=lin.predict([X_train[i]])
    temp=(np.abs(Val-(Y_train[i]))/(Y_train[i]))*100
    per.append(temp)
avgp=sum(per)/len(per)
print("average percentage error in price prediction of train set is: {}% ".format(avgp[0]))
```

```python
# calculating average percentage error in predicted price for test set
per1=[]
import math
for i in range(Y_test.shape[0]):
    Val=lin.predict([X_test[i]])
    temp=(np.abs(Val-(Y_test[i]))/(Y_test[i]))*100
    per1.append(temp)
avgp1=sum(per1)/len(per1)
print("average percentage error in price prediction of test set is: {}% ".format(avgp1[0]))
```

```python
#calculation of R2 and Root mean square error for train set for checking performance of model 
Y_train_predict = lin.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train, Y_train_predict)))
r=r2_score(Y_train,Y_train_predict)
print("value of R2 for train set is: ",r)
print("Root mean square value for train set is: ",rmse)
```

```python
#calculation of R2 and Root mean square error for test set for checking performance of model
Y_test_predict = lin.predict(X_test)
rmse2 = (np.sqrt(mean_squared_error(Y_test, Y_test_predict)))
r2 = r2_score(Y_test, Y_test_predict)
print("value of R2 for test set is: ",r2)
print("Root mean square value for test set is: ",rmse2)
```
